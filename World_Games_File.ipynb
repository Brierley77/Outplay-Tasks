```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

import matplotlib
matplotlib.rcParams['axes.grid'] = True # show gridlines by default
%matplotlib inline

from pandas import *
```

```python
# Setting LOCATION as a file to be read. 
```

```python
LOCATION = 'UK_Games.csv'
```

```python
# Setting UK_Games as the file, and asking for the columns 'Commodity Code' and 'Reporter Code' to be read as strings.
# I ask for my data UK_Games to show me the first 5 rows of my newly created data set. 
```

```python
UK_Games = read_csv(LOCATION, dtype={'Commodity Code':str, 'Reporter Code':str, 'Year': str})
UK_Games.head()
```

```python
# Set new columns for my data set, so I can ignore columns I do not need. 
```

```python
COLUMNS = ['Year', 'Trade Flow', 'Reporter', 'Partner', 'Commodity', 'Commodity Code', 'Trade Value (US$)']
UK_Games = UK_Games[COLUMNS]
```

```python
# Check to see if this has worked. 
```

```python
UK_Games.head()
```

```python
# Seperated and created new data sets just focusing on imports and exports. 
```

```python
Games_imports = UK_Games[UK_Games['Trade Flow'] == 'Import']
Games_exports = UK_Games[UK_Games['Trade Flow'] == 'Export']
```

```python
Games_imports.head()
```

```python
Games_exports.head()
```

```python
# Seperated import data for the year 2008
```

```python
UKImportsIn2008 = Games_imports[Games_imports['Year'] == '2008']
```

```python
UKImportsIn2008.head()
```

```python
# Seeing who we imported the most from in terms of video game technology in 2008. 
# Table shows the top 10 importing countries to the UK in descending order in US $. 
```

```python
UKImportsIn2008.sort_values('Trade Value (US$)', ascending=False).head(10)
```

```python
# Seeing who exported the most in terms of video game technology in 2002.
# Table shows the top 10 exporting countries from the UK in descending order in US $.
```

```python
UKExportsIn2002 = Games_exports[Games_exports['Year'] == '2002']
UKExportsIn2002.sort_values('Trade Value (US$)', ascending = False).head(10)
```

```python
# Grouping data sets. 
```

```python
groups = UK_Games.groupby('Trade Flow')
```

```python
groups.get_group('Import').head()
```

```python
# Grouped by a specific Country. 
```

```python
groups = UK_Games.groupby('Reporter')
groups.get_group('France').head()
```

```python
# Specified for columns reporter and trade flow to group and set this = to groups

# Then informed groups to only output france and import. 
```

```python
GROUPING_REPORTERFLOW = ['Reporter', 'Trade Flow']
groups = UK_Games.groupby (GROUPING_REPORTERFLOW)

GROUP_REPORTERFLOW = ('France', 'Import')
groups.get_group(GROUP_REPORTERFLOW)
```

```python
# Grouped by counrty to find out their import totals. 
# Only grabbed first 10. 
```

```python
Games_imports.groupby('Reporter')['Trade Value (US$)'].aggregate(sum).head(10)
```

```python
# Grouped by counrty to find out their export totals. 
# Only grabbed first 10. 
```

```python
Games_exports.groupby('Reporter')['Trade Value (US$)'].aggregate(sum).head(10)
```

```python
# Finding top 5 importing counrties and their total value over the year 2014 - 2002.
```

```python
Games_imports_totals = Games_imports.groupby('Reporter')[['Trade Value (US$)']].aggregate(sum)
Games_imports_totals.sort_values('Trade Value (US$)', ascending = False).head()
```

```python
# Finding top 5 exporting counrties and their total value over the year 2014 - 2002.
```

```python
Games_exports_totals = Games_exports.groupby('Reporter')[['Trade Value (US$)']].aggregate(sum)
Games_exports_totals.sort_values('Trade Value (US$)', ascending = False).head()
```

```python
# Plotting a graph showing the top 5 imports of video game technology and approx how much in the  100 millions 
```

```python
Top_5_Imports = Games_imports.groupby('Reporter')[['Trade Value (US$)']].aggregate(sum)
Top_5_Imports.sort_values('Trade Value (US$)', ascending =False).head().plot(kind='barh')
```

```python
# PLotting a graph showing the top 5 expors of video game technology and apporx how much in the billions 
```

```python
Top_5_Exports = Games_exports.groupby('Reporter')[['Trade Value (US$)']].aggregate(sum)
Top_5_Exports.sort_values('Trade Value (US$)', ascending = False).head().plot(kind='barh')
```
